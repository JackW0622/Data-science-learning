First, we slice the DEM data file into various tiles with slight overlapping in order to clean the data furthermore for future machine learning. The overlap is designed so that the context of each tile in relation to its surroundings can be counted into the learning process, generating more realistic DEM file for output.

```python
import os
import math
import rasterio
from rasterio.windows import Window
from rasterio.transform import Affine

in_dem_path = "/ASTGTMV003_N00W065_dem.tif"
out_dir = "/bin/dem sliced file"
tile_size = 512
overlap = 32

os.makedirs(out_dir, exist_ok=True)

with rasterio.open(in_dem_path) as src:
    width = src.width
    height = src.height
    profile = src.profile.copy()

    # Compute number of tiles in each direction
    stride = tile_size - overlap
    n_tiles_x = math.ceil((width - overlap) / stride)
    n_tiles_y = math.ceil((height - overlap) / stride)

    for i in range(n_tiles_x):
        for j in range(n_tiles_y):
            x_off = i * stride
            y_off = j * stride

            # If window goes beyond raster, clamp
            win_width = min(tile_size, width  - x_off)
            win_height= min(tile_size, height - y_off)

            window = Window(x_off, y_off, win_width, win_height)

            # Compute the transform for this window
            transform = src.window_transform(window)

            # Update profile
            profile.update({
                "height": win_height,
                "width":  win_width,
                "transform": transform
            })

            out_name = f"tile_{i:03d}_{j:03d}.tif"
            out_path = os.path.join(out_dir, out_name)

            # Read and write
            data = src.read(window=window)
            with rasterio.open(out_path, "w", **profile) as dst:
                dst.write(data)
```

We need to determine the power-law slope(β) in order to generate realistic amazonian landscape using Perlin Noise algorithm.

```python
import numpy as np
import rasterio
from numpy.fft import fft2, fftshift
from scipy.stats import linregress
import matplotlib.pyplot as plt

# 1) Load + mask no-data
with rasterio.open("/tile_000_001.tif") as src:
    dem = src.read(1).astype(float)
    nodata = src.nodata

# Replace NaNs & nodata with the overall mean
dem[np.isnan(dem) | (dem == nodata)] = np.nanmean(dem)

# Subtract the global mean (no planar detrend)
dem_centered = dem - np.nanmean(dem)

# 2) Window taper
nrows, ncols = dem_centered.shape
winx = np.hanning(ncols)
winy = np.hanning(nrows)
window2d = np.sqrt(np.outer(winy, winx))
dem_win = dem_centered * window2d

# 3) FFT + PSD
F   = fftshift(fft2(dem_win))
psd = np.abs(F)**2

# 4) Radial binning
kx = np.fft.fftfreq(ncols).reshape(1, ncols)
ky = np.fft.fftfreq(nrows).reshape(nrows, 1)
k  = np.sqrt(kx**2 + ky**2)
k_flat, psd_flat = k.ravel(), psd.ravel()
mask = psd_flat > 0
k_flat, psd_flat = k_flat[mask], psd_flat[mask]

k_bins = np.linspace(0, k_flat.max(), 100)
bin_idx = np.digitize(k_flat, k_bins)
k_centers = 0.5*(k_bins[:-1] + k_bins[1:])
psd_radial = np.array([
    psd_flat[bin_idx == i].mean() if np.any(bin_idx == i) else np.nan
    for i in range(1, len(k_bins))
])

valid = ~np.isnan(psd_radial)
k_centers = k_centers[valid]
psd_radial = psd_radial[valid]

# 5) Select fitting range
k_min = 2.0 / ncols
k_max = 1.0 / 8.0
fit_mask = (k_centers > k_min) & (k_centers < k_max)

# 6) Fit in log–log
if np.sum(fit_mask) >= 2:
    logk  = np.log10(k_centers[fit_mask])
    logps = np.log10(psd_radial[fit_mask])
    slope, intercept, *_ = linregress(logk, logps)
    beta = -slope
    print(f"Estimated β = {beta:.2f}")
else:
    print("Not enough data points in the selected range for regression.")
    logk = logps = np.array([])

# 7) Plot log–log fit
plt.figure(figsize=(6,4))
if logk.size:
    plt.scatter(logk, logps, s=20, alpha=0.6, label="data")
    plt.plot(logk, intercept + slope*logk, 'k--',
             label=f"slope = {slope:.2f}")
plt.xlabel("log10(k)"); plt.ylabel("log10(PSD)")
plt.legend()
plt.title("Log–log fit for β")
plt.show()

# 8) Plot radial PSD with fit window
plt.figure(figsize=(6,4))
plt.scatter(k_centers, psd_radial, s=10, alpha=0.5, label="all bins")
plt.scatter(k_centers[fit_mask], psd_radial[fit_mask],
            c='C1', s=20, label="fit bins")
plt.xscale('log'); plt.yscale('log')
plt.axvline(k_min, color='grey', ls='--')
plt.axvline(k_max, color='grey', ls='--')
plt.xlabel("wavenumber k"); plt.ylabel("PSD(k)")
plt.legend()
plt.title("Radial PSD with fit window")
plt.show()
```

The beta value we obtained is too low, one idea is that the tile I chose would be representing such a floodplain which no clear trend can be identified after the cleaning and detrending. Another idea is to swap the Hann method into Tukey, a smoother and softer method to try again for better results.


```python
import numpy as np
import rasterio
from numpy.fft import fft2, fftshift
from scipy.stats import linregress
import matplotlib.pyplot as plt

# 1) Load + mask no-data
with rasterio.open("/tile_000_001.tif") as src:
    dem = src.read(1).astype(float)
    nodata = src.nodata

# Replace NaNs & nodata with the overall mean
dem[np.isnan(dem) | (dem == nodata)] = np.nanmean(dem)

# Subtract the global mean (no planar detrend)
dem_centered = dem - np.nanmean(dem)

# 2) Window taper
def tukey(N, alpha=0.1):
    """
    Pure-NumPy Tukey window.
    N     : window length
    alpha : fraction of the window inside the cosine tapered regions
    """
    if alpha <= 0:
        return np.ones(N)
    if alpha >= 1:
        # falls back to a Hann if alpha=1
        return np.hanning(N)

    # normalized indices in [0,1]
    n = np.linspace(0, 1, N)
    w = np.ones(N)

    # first taper region
    left = n < alpha/2
    w[left] = 0.5 * (1 + np.cos(2*np.pi*(n[left]/alpha - 0.5)))

    # last taper region
    right = n > 1 - alpha/2
    w[right] = 0.5 * (1 + np.cos(2*np.pi*((n[right] - 1)/alpha + 0.5)))

    return w

alpha = 0.1   # 10% taper on each side
winx = tukey(ncols, alpha=alpha)
winy = tukey(nrows, alpha=alpha)
window2d = np.outer(winy, winx)
dem_win = dem_centered * window2d

# 3) FFT + PSD
F   = fftshift(fft2(dem_win))
psd = np.abs(F)**2

# 4) Radial binning
kx = np.fft.fftfreq(ncols).reshape(1, ncols)
ky = np.fft.fftfreq(nrows).reshape(nrows, 1)
k  = np.sqrt(kx**2 + ky**2)
k_flat, psd_flat = k.ravel(), psd.ravel()
mask = psd_flat > 0
k_flat, psd_flat = k_flat[mask], psd_flat[mask]

k_bins = np.linspace(0, k_flat.max(), 100)
bin_idx = np.digitize(k_flat, k_bins)
k_centers = 0.5*(k_bins[:-1] + k_bins[1:])
psd_radial = np.array([
    psd_flat[bin_idx == i].mean() if np.any(bin_idx == i) else np.nan
    for i in range(1, len(k_bins))
])

valid = ~np.isnan(psd_radial)
k_centers = k_centers[valid]
psd_radial = psd_radial[valid]

# 5) Select fitting range
k_min = 2.0 / ncols
k_max = 1.0 / 8.0
fit_mask = (k_centers > k_min) & (k_centers < k_max)

# 6) Fit in log–log
if np.sum(fit_mask) >= 2:
    logk  = np.log10(k_centers[fit_mask])
    logps = np.log10(psd_radial[fit_mask])
    slope, intercept, *_ = linregress(logk, logps)
    beta = -slope
    print(f"Estimated β = {beta:.2f}")
else:
    print("Not enough data points in the selected range for regression.")
    logk = logps = np.array([])

# 7) Plot log–log fit
plt.figure(figsize=(6,4))
if logk.size:
    plt.scatter(logk, logps, s=20, alpha=0.6, label="data")
    plt.plot(logk, intercept + slope*logk, 'k--',
             label=f"slope = {slope:.2f}")
plt.xlabel("log10(k)"); plt.ylabel("log10(PSD)")
plt.legend()
plt.title("Log–log fit for β")
plt.show()

# 8) Plot radial PSD with fit window
plt.figure(figsize=(6,4))
plt.scatter(k_centers, psd_radial, s=10, alpha=0.5, label="all bins")
plt.scatter(k_centers[fit_mask], psd_radial[fit_mask],
            c='C1', s=20, label="fit bins")
plt.xscale('log'); plt.yscale('log')
plt.axvline(k_min, color='grey', ls='--')
plt.axvline(k_max, color='grey', ls='--')
plt.xlabel("wavenumber k"); plt.ylabel("PSD(k)")
plt.legend()
plt.title("Radial PSD with fit window")
plt.show()
```
